```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
```

## Calculating Engagement Score

```python
df=pd.read_csv("./reports/report.csv",index_col=None)

y=df["Result"]

ecount=0
ncount=0

y=y.to_numpy()

for i in range(len(y)):
    if y[i] ==1:
        ecount=ecount+1
    else:
        ncount=ncount+1

escore=ecount/(ecount+ncount)*100
```

```python
detail=pd.read_csv("./uploaded/details/details.csv",index_col=None)
sname=detail["Name"]
smarks=detail["Marks"]
stime=detail["Time"]
sname=sname[0]
smarks=smarks[0]
stime=stime[0]
```

```python
fig = plt.figure(figsize = (20, 4))
# creating the bar plot
values=list(range(1,len(y)+1))
plt.bar(values, y, color ='purple',width = 0.4)
plt.xlabel("Frames")
plt.ylabel("Engagement")
plt.title(sname+"\'s Engagement")
plt.savefig("./reports/EngagementGraph.png",dpi=1000,bbox_inches="tight")
plt.show()
```

## Percentage Engagement

```python
# Data to plot
labels = 'Engaged','Not Engaged'
sizes = [ecount,ncount]
colors = ['yellowgreen', 'red']

plt.pie(sizes,labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
plt.legend(labels, loc="best")
plt.axis('equal')
plt.title(sname+"\'s Engagement Score")
plt.savefig("./reports/EngagementScore.png",dpi=1000,bbox_inches="tight")
plt.show()
```

## Quiz Score

```python
# Data to plot
labels = 'Correct','Wrong'
sizes = [smarks,10-smarks]
colors = ['skyblue', 'red']

plt.pie(sizes,labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=90)
plt.legend(labels, loc="best")
plt.axis('equal')
plt.title(sname+"\'s Quiz Score")
plt.savefig("./reports/QuizScore.png",dpi=1000,bbox_inches="tight")
plt.show()
```

## Calculating Overall Score

```python
rscore=0
qscore=smarks/10*100
if(stime==0):
    rscore=escore*0.6+qscore*0.4
elif(stime==1):
    rscore=escore*0.7+qscore*0.3
elif(stime==2):
    rscore=escore*0.75+qscore*0.25
    
# Data to plot
labels = 'Engaged','Not Engaged'
sizes = [rscore,100-rscore]
colors = ['magenta', 'purple']

plt.pie(sizes,labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=90)
plt.legend(labels, loc="best")
plt.axis('equal')
plt.title(sname+"\'s Overall Engagement")
plt.savefig("./reports/OverallScore.png",dpi=1000,bbox_inches="tight")
plt.show()
```

## Result

```python
strReport=""
if(rscore<=25):
    strReport="Not Engaged"
elif(rscore>25 and rscore<=50):
    strReport="Partially Engaged"
elif(rscore>50 and rscore<=75):
    strReport="Engaged"
else:
    strReport="Fully Engaged"
print(sname+"\'s Overall Score")
print(strReport)
    
```
